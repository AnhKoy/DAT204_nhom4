```python
import pandas as pd
import re
from collections import Counter
```

```python
data = pd.read_csv('/content/dataset_linkedin-jobs(full)2025-01-19_01-38-15-431.csv')

# Inspect the first few rows
print(data.head())
```

```python
skills = [
    'python', 'sql', 'excel', 'r studio', 'power bi', 'powerbi', 'tableau', 'machine learning',
    'deep learning', 'aws', 'azure', 'gcp', 'hadoop', 'spark', 'data visualization',
    'data analytics', 'statistics', 'numpy', 'pandas', 'scikit-learn', 'nlp', 'big data',

    'google analytics', 'data warehouse', 'snowflake', 'brick', 'mysql', 'sql server', 'postgres', 'olap', 'dax', 'power platform', 'power automate', 'database'
]
```

```python
def preprocess_text(text):
    text = text.lower()  # Convert to lowercase
    text = re.sub(r'[^\w\s]', '', text)  # Remove punctuation
    return text

# Apply preprocessing to job descriptions
data['clean_description'] = data['description_text'].apply(preprocess_text)
```

```python
def extract_skills(text, skills_list):
    found_skills = [skill for skill in skills_list if skill in text]
    return found_skills

# Extract skills from each job description
data['skills'] = data['clean_description'].apply(lambda x: extract_skills(x, skills))
```

```python
# Làm sạch cột "skills" hoàn chỉnh
import ast

def clean_skills_column(skill_str):
    """Chuẩn hóa danh sách kỹ năng: chuyển thành chữ thường, loại bỏ khoảng trắng và trùng lặp."""
    if isinstance(skill_str, str):  # Kiểm tra nếu là chuỗi
        try:
            skills_list = ast.literal_eval(skill_str) if skill_str.startswith("[") else skill_str.split(",")
            if isinstance(skills_list, list):
                skills_list = [skill.strip().lower() for skill in skills_list if skill.strip()]
                return ", ".join(sorted(set(skills_list)))  # Loại bỏ trùng lặp & sắp xếp gọn gàng
        except:
            return ""
    return ""

# Áp dụng làm sạch vào cột "skills"
df_clean["skills"] = df_clean["skills"].apply(clean_skills_column)

# Kiểm tra lại kết quả
df_clean["skills"].head()
```

```python
import re
import pandas as pd

# Hàm tách mức lương thấp nhất và cao nhất
def extract_salary(salary):
    if isinstance(salary, str):  # Kiểm tra nếu giá trị là chuỗi
        # Loại bỏ ký hiệu $, dấu phẩy, khoảng trắng và đơn vị thời gian
        salary = re.sub(r'[\$,]', '', salary)
        salary = re.sub(r'(/yr|/mo|/hr)', '', salary)

        # Tìm tất cả số trong chuỗi
        matches = re.findall(r'\d+\.\d+|\d+', salary)  # Bắt cả số thập phân và số nguyên

        if len(matches) == 2:  # Trường hợp có khoảng lương
            return float(matches[0]), float(matches[1])
        elif len(matches) == 1:  # Trường hợp chỉ có một mức lương
            value = float(matches[0])
            return value, value  # Gán cả hai giá trị như nhau

    return None, None  # Trường hợp NaN hoặc không hợp lệ

# Áp dụng hàm cho cột base_salary
df_clean[['min_salary', 'max_salary']] = df_clean['base_salary'].apply(
    lambda x: pd.Series(extract_salary(x))
)

# Xóa cột cũ "base_salary"
df_clean.drop(columns=['base_salary'], inplace=True)

# Kiểm tra kết quả
df_clean[['min_salary', 'max_salary']].head()
```
